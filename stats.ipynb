```python
''' 
Q.1 What is the meaning of six sigma in statistics?  Give proper example 
Ans. 
Six Sigma is a set of techniques and tools used for process improvement. It was originally developed by Motorola in the 1980s and later popularized by companies like General Electric. The term "Six Sigma" refers to a statistical concept that measures how far a process deviates from perfection.

In statistics, "Sigma" represents the standard deviation, which is a measure of the dispersion or variability in a set of data. The goal of Six Sigma is to minimize variability and defects in a process to achieve near-perfect quality.

The Six Sigma methodology typically follows a structured approach known as DMAIC, which stands for Define, Measure, Analyze, Improve, and Control. This approach involves identifying the problem, measuring the current performance, analyzing the root causes of defects, implementing improvements, and establishing controls to sustain the improvements.
Here's an example to illustrate Six Sigma:
Let's consider a manufacturing process that produces widgets. The length of each widget is critical, and any deviation from the target length results in defects. The target length of the widgets is 10 centimeters.
Define: The problem is defined as excessive variability in widget length, leading to defects and customer dissatisfaction.
Measure: Data is collected on the lengths of widgets produced over a period of time. A histogram or other statistical tools are used to analyze the data and determine the current performance of the process.
Analyze: Statistical analysis is performed to identify the root causes of variability in widget length. Possible causes may include machine calibration issues, operator error, or variability in raw materials.
Improve: Based on the analysis, improvements are implemented to reduce variability in the process. This may involve recalibrating machines, providing additional training to operators, or improving supplier quality.
Control: Control measures are established to ensure that the improvements are sustained over time. Statistical process control (SPC) charts are used to monitor the process and detect any deviations from the desired performance.
By following the DMAIC approach and applying statistical tools and techniques, the Six Sigma methodology aims to achieve significant quality improvements and cost savings in various industries.


'''
```

```python
'''
Q.3 What is the meaning of the five-number summary in Statistics? Give proper example
Ans. The five numbers in the five-number summary are:

Minimum: The smallest value in the dataset.
First quartile (Q1): The value below which 25% of the data falls.
Median (Q2 or the second quartile): The middle value of the dataset when it is sorted in ascending order. It divides the dataset into two equal halves.
Third quartile (Q3): The value below which 75% of the data falls.
Maximum: The largest value in the dataset.

Consider the following dataset representing the ages of a group of individuals:
20,25,30,35,40,45,50,55,60,65

To find the five-number summary:

Minimum: 20
First quartile (Q1): 25
Median (Q2): 42.5
Third quartile (Q3): 55
Maximum: 65
So, the five-number summary for this dataset is: 
20,25,42.5,55,65.


'''
```

```python
'''
Q.4 What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook
Ans. 
'''

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
hours_studied = [2, 3, 4, 5, 6, 7, 8, 9, 10]
exam_scores = [60, 65, 70, 75, 80, 85, 90, 95, 100]

# Create a DataFrame
data = pd.DataFrame({'Hours Studied': hours_studied, 'Exam Scores': exam_scores})

# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(data['Hours Studied'], data['Exam Scores'])[0, 1]
print("Correlation coefficient:", correlation_coefficient)

# Plotting
sns.scatterplot(data=data, x='Hours Studied', y='Exam Scores')
plt.title('Hours Studied vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.grid(True)
plt.show()
```
